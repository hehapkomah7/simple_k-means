```python
import numpy as np
import matplotlib.pyplot as plt
import random

n_features = 2  # two features belonged to coordinate axes
n_samples = random.randint(100, 300)  # randomly generated number of samples
n_clusters = random.randint(2, 10)  # randomly generated number of clusters
iterations = random.randint(5, 100)  # randomly generated number of iterations

F = n_features
S = n_samples
C = n_clusters
I = iterations
II = 0

print('number of Samples:', S)
print('number of Features:', F)
print('number of Clusters:', C)
print('number of Iterations:', I)

x = np.random.randn(n_samples, n_features)
y = np.zeros(n_samples)  # creating random positions
for t in range(II):
    if t == 0:
        index_ = np.random.choice(range(n_samples), n_clusters, replace=False)  #
        mean = x[index_]
    else:
        for k in range(n_clusters):
            mean[k] = np.mean(x[y == k], axis=0)
    for i in range(n_samples):
        dist = np.sum((mean - x[i]) ** 2, axis=1)
        pred = np.argmin(dist)
        y[i] = pred  # Lloyd's algorithm

for k in range(n_clusters):
    fig1 = plt.scatter(x[y == k, 0], x[y == k, 1])
    plt.title('Before Iterations')
plt.show()


x = np.random.randn(n_samples, n_features)
y = np.zeros(n_samples)  # creating random positions
for t in range(iterations):
    if t == 0:
        index_ = np.random.choice(range(n_samples), n_clusters, replace=False)  #
        mean = x[index_]
    else:
        for k in range(n_clusters):
            mean[k] = np.mean(x[y == k], axis=0)
    for i in range(n_samples):
        dist = np.sum((mean - x[i]) ** 2, axis=1)
        pred = np.argmin(dist)
        y[i] = pred  # Lloyd's algorithm

for k in range(n_clusters):
    fig2 = plt.scatter(x[y == k, 0], x[y == k, 1])
    plt.title('After Iterations')
plt.show()
```
